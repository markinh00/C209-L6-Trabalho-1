```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
def get_face(img, min_size=20):
    (l, c, p) = img.shape
    img_gray = ((0.2126 * img[:, :, 0]) + (0.7152 * img[:, :, 1]) + (0.0722 * img[:, :, 2])).astype(np.uint8)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
      
    data = cv2.CascadeClassifier('raw.githubusercontent.com_opencv_opencv_master_data_haarcascades_haarcascade_frontalface_default.xml')
    found = data.detectMultiScale(img_gray, minSize=(min_size, min_size))
    amount_found = len(found)
    
    if amount_found != 0:
        for (x, y, width, height) in found:
            cv2.rectangle(img_rgb, (x, y), 
                          (x + height, y + width), 
                          (0, 255, 0), 5)

    return img_rgb
```

```python
img1 = cv2.imread("images/b3b7ceccdc942f40ea39545cb6ba7a90.jpg")
img2 = cv2.imread("images/best-face-oil.webp")
img3 = cv2.imread("images/16529839176286886de359a_1652983917_3x4_md.jpg")
img4 = cv2.imread("images/15ea57e1305ad7803832ab71c012fca1.jpg")

min_size = 150
img_rgb1 = get_face(img1, min_size)
img_rgb2 = get_face(img2, min_size)
img_rgb3 = get_face(img3, min_size)
img_rgb4 = get_face(img4, min_size)

plt.figure(figsize=(8,8))
plt.subplot(2, 2, 1)
plt.imshow(img_rgb1)
plt.subplot(2, 2, 2)
plt.imshow(img_rgb2)
plt.subplot(2, 2, 3)
plt.imshow(img_rgb3)
plt.subplot(2, 2, 4)
plt.imshow(img_rgb4)
plt.show()
```
